Making Changes to append
We need to make some changes to this dataframe in order to append to the final dataframe

```python
#rename as same as final dataframe
test.rename(columns={"Base Price IN ₹":"Base Price"},inplace=True)
```

```python
#rename as same as final dataframe
test.rename(columns={"Players":"Player"},inplace=True)
```

```python
#dropping unwanted column
test.drop("Base Price IN $",axis=1,inplace=True)
```

```python
#creating new columns with null values as same as final dataframe
test["COST IN ₹ (CR.)"]= np.nan
```

```python
#creating new columns with null values as same as final dataframe
test["Cost IN $ (000)"]= np.nan
```

```python
#Adding Team name as Unsold
test["Team"]="Unsold"
```
